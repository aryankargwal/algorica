```python
# Importing Libraries
import numpy as np 
import math 
import matplotlib.pyplot as plt
```

```python
# Defining Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Defining Variables
x = np.linspace(-10,10,100)
y = x * sigmoid(x)
```

```python
# Plotting the Function
plt.plot(x,y)
plt.show()
```
